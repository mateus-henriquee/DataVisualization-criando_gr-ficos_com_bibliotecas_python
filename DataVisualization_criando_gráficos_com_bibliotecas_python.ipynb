# Cáp 1. Conhecendo a biblioteca Matplotlib 🔍

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/dataset.csv")
# https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv <-- Link csv
```

```python
df
```

```python
df.info() # <-- Visualizar tipo de dados armazenados
```

```python
# Tarefa 1: Analizar as tendências de imigração do Brasil em um determinado período
df.set_index("País", inplace=True) # <-- Criando index como país
```

```python
anos = list(map(str, range(1980, 2014))) # <-- Definindo lista de 1980 até 2014
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos] # <-- "loc['']" utilizada para pegar rotulos especificos de um dataframe
```

```python
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()} # <-- Dicionario para criar um novo dataframe dos dados do brasil (tolist() --> converte uma series em uma lista)
dados_brasil = pd.DataFrame(brasil_dict) # <-- Criando dataframe
```

```python
dados_brasil
```

```python
# -- BONUS --
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5)) # <-- Alterando tamanho do plot/grafico
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) # <-- Selecionando dados do grafico como ano no eixo 'X' e imigrante no eixo 'Y'
plt.title('Imigração do Brasil para o Canadá') # <-- Informa titulo do plot
plt.xlabel('Ano(s)') # <-- Informa um rotulo para o eixo 'X'
plt.ylabel('Número de Imigrantes') # <-- Informa um rotulo para o eixo 'Y'
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) # <-- Filtrando os anos que serão exibidos no grafico
# plt.yticks([500, 1000, 1500, 2000, 2500, 3000])  # <-- Filtrando valor de imigrandos do grafico
plt.show() # <-- Ao utilizar, a saida é mais limpa
```

```python
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
```

```python
df_comparacao = df_comparacao.T # <-- .T é transpose, trnasforma linhas em colunas
df_comparacao.head()
```

```python
plt.plot(df_comparacao['Brasil'],label = 'Brasil')
plt.plot(df_comparacao['Argentina'],label ='Argentina')
plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

# Cáp 2. Criando figuras com Matplotlib 📈

```python
fig, ax = plt.subplots(figsize=(8,4)) # <-- Criando subplots, fig = figura em branco; ax = "axes" no portugues "eixos"
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá\n 1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de Imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # <-- Ao inves de selecionar anos um de cada vez, 'xaxis.set_major_locator' modificador do eixo 'X'; 'MultipleLocator()' = Altera a frequencia do eixo selecionado
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n 1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n 1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')

plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.5)

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colombia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de Imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)


plt.show()
```

# Cáp. 3 Customizando com Matplotlib 📝

```python
fig, ax = plt.subplots(figsize=(8,4)) # <-- Criando subplots, fig = figura em branco; ax = "axes" no portugues "eixos"
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=2, color='dodgerblue') # <-- lw aumenta a "grossura" da linha do grafico; marker="o" adciona marcadores na lina do grafico *atenção para não sobrecarregar a visualização do usuario
ax.set_title('Imigração do Brasil para o Canadá\n 1980 a 2013', fontsize=14, loc='Left') # <-- Fontsize aumenta tamnho da fonte dos titulos; <loc> Se utiliza para alterar posição do titulo
ax.set_xlabel('Ano' , fontsize=14)
ax.set_ylabel('Número de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12) # <-- Label size para alterar tamanho da fonto das label's
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # <-- Ao inves de selecionar anos um de cada vez, 'xaxis.set_major_locator' modificador do eixo 'X'; 'MultipleLocator()' = Altera a frequencia do eixo selecionado
# plt.grid(linestyle='--') <-- Adiciona uma grade ao fundo branco para guiar a visualização dos valores mostrados
ax.spines['top'].set_visible(False) # <-- Removendo a linha do topo
ax.spines['right'].set_visible(False) # <-- Removendo a linha da direita


fig.savefig('imigracao_brasil_canada.png', transparent=False, dpi=300, bbox_inches='tight') # <-- <-- Salvando png da imagem; dpi vai alterar o tamanho da imagem em px; bbox ajusta automaticamente os limites da imagem

plt.show()


# -- DICA --
# A visualização das pessoas funcionam com formato em Z sendo olhando do topo da esquerda para direita.
# Ideal: Colocar titulos no topo da esquerda para facilitar leitura do usuario
```

```python
america_sul = df.query('Região == "América do Sul"') # <-- Criando dataframe para apenas paises da américa do sul
```

```python
cores = ['royalblue',
         'orange',
         'forestgreen',
         'orchid',
         'purple',
         'brown',
         'slateblue',
         'gray',
         'olive',
         'navy',
         'teal',
         'tomato']
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.bar(america_sul.index, america_sul['Total'], color=cores) # <-- ax.bar é o grafico de barrras
ax.set_title('Imigração da América do Sul para o Canadá\n 1980 a 2013', loc='Left', fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul.index, america_sul['Total'], color=cores) # <-- ax.barh transforma em um grafico de barras na horizontal
ax.set_title('Imigração da América do Sul para o Canadá\n 1980 a 2013', loc='Left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
america_sul_ordenada = america_sul.sort_values('Total', ascending=True) # <-- Ordenando o total de imigrantes com ascending=True
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_ordenada.index, america_sul_ordenada['Total'], color=cores) # <-- Troca de dataframe para colocar as barras em ordem
ax.set_title('Imigração da América do Sul para o Canadá\n 1980 a 2013', loc='Left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
# Ideia: DESTACAR O BRASIL DENTRE OS PAISES DA AMERICA DO SUL
```

```python
cores = [] # <-- Deixando lista de cores vazias para fazer alterações

for pais in america_sul_ordenada.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

# Loop para destacar a cor caso pais seja Brasil
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_ordenada.index, america_sul_ordenada['Total'], color=cores) # <-- Troca de dataframe para colocar as barras em ordem
ax.set_title('Imigração da América do Sul para o Canadá de 1980 a 2013', loc='Left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)


for i, v in enumerate(america_sul_ordenada['Total']): # <-- Loop para colocar o valor 'Total' de imigrantes de cada pais
  ax.text(v + 100, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False) # <-- Retira linhas/caixa do grafico deixa somente os valores
ax.get_xaxis().set_visible(False) # <-- Retira o eixo X por completo
ax.tick_params(axis='both', which='both', length=0) # <-- Remove alguns parametros padrões do plot como exemplo Brasil com um "-" ao lado

fig.savefig('imigracao_america_sul_brasil_destacado.png', transparent=False, dpi=500, bbox_inches='tight') # <-- Salvando imagem; dpi vai alterar o tamanho da imagem em px; bbox ajusta automaticamente os limites da imagem

plt.show()
```

# Cáp. 4 Conhecendo a biblioteca "Seaborn" 🕵️

```python
import seaborn as sns # <-- Biblioteca de gráficos e estilização
```

```python
sns.set_theme() # <-- Tema padrão; Ao utilizar TODOS os graficos terão este tema
sns.set_style()
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10) # <-- Estratégia: .head(10) já seleciona automaticamente a ordem decrescente do total de imigrantes dos paises
```

```python
top_10
```

```python
sns.barplot(data=top_10, x = top_10.index, y = 'Total')
```

```python
sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h') # <-- Plotando grafico de barras horizontal; orient=h/v muda o sentido do grafico
```

```python
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')
ax.set(title='Paises com maior imigração para o Canadá',
       xlabel='Número de imigrantes',
       ylabel='')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='Left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
plt.show()
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)

  ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='Left', fontsize=16)
  ax.set_xlabel('Número de imigrantes', fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

```python
gerar_grafico_paleta('Blues')
```

```python
gerar_grafico_paleta('Blues_r') # <-- _r deixa na ordem decrescente de cores
```

```python
gerar_grafico_paleta('rocket')
```

```python
gerar_grafico_paleta('coolwarm') # <-- Paleta ideal para identificar o "meio" do grafico
```

```python
gerar_grafico_paleta('tab10') # <-- Ideal para destinguir os dados e também para nenhuma cor se destacar da outra
```

```python
sns.set_theme(style='dark') # <-- Sem o grid e com o fundo um pouco mais escruo, bom quando tiver muitas cores utilizadas
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='white') # <-- Tema com fundo branco
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks') # <-- Tema que adiciona "-" nos itens
gerar_grafico_paleta('tab10')
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette='tab10')

ax.set_title('Países com maior imigração para o Canadá\n 1980 a 2013', loc='Left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
sns.despine() # <-- Função para remover a "caixa"/"bordas" do grafico

plt.show()
```

# Cáp 5. Gráficos interativos com Plotly 📶

```python
import plotly.express as px # <-- Biblioteca que permite criar graficos interativos com poucas linhas de código
```

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes', title='Dados de imigração do Brasil para o Candá de 1980 até 2013')

fig.update_traces(line_color='green', line_width=3) # <-- Altera a coloração da linha do grafico
fig.update_layout(width=1000, height=500, # <-- Função para atualizar o grafico; width (largura); height (altura);
                  xaxis={'tickangle':-40}, # <-- 'tickanlge' gira o texto em º
                  font_family='Arial', # <-- Altera a fonte do texto
                  font_size=14, # <-- Altera o tamanho do texto
                  font_color='grey', # <-- Altera a cor do texto
                  title_font_color='black', # <-- Altera a cor do titulo
                  title_font_size=22, # <-- Altera a fonte do titulo
                  xaxis_title='Ano', # <-- Adiciona titulo ao eixo 'X'
                  yaxis_title='Número de imigrantes') # <-- Adiciona titulo ao eixo 'Y'
fig.show()
```
